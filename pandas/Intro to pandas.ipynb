```python
import pandas as pd

cols=['Name', 'Subject', 'Marks','Grade']

student_data=pd.DataFrame(columns=cols)
student_data
```

```python
#task1

data={'Name': ['Alice','Bob','Charlie', 'Alice', 'Bob'],
      'Subject':['Math', 'Science', 'History', 'Science', 'Math'],
      'Marks':[85,90,78,88,92],
      'Grade':['B', 'A', 'C','B','A']}

student_data=pd.DataFrame(data)
student_data
```

```python
#task2

student_data['Bonus Marks']=0.1*student_data['Marks']
student_data
```

```python
#task3

grouped=student_data.groupby('Name')

avg_marks=grouped['Marks'].mean()
marks_data=pd.DataFrame(avg_marks)
marks_data.rename(columns={'Marks': 'Average Marks'}, inplace=True)

marks_data
```

```python
total_bonus_marks=grouped['Bonus Marks'].sum()
bonus_data=pd.DataFrame(total_bonus_marks)
bonus_data.rename(columns={'Bonus Marks':'Total Bonus Marks'},inplace=True)

bonus_data
```

```python
result = pd.DataFrame({
    'Name': avg_marks.index,
    'Average Marks': avg_marks.values,
    'Total Bonus Marks': total_bonus_marks.values
})

result =result.reset_index(drop=True)

result_data=pd.DataFrame(result)
result_data
```
